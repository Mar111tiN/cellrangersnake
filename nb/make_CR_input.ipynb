### get features for running on snakemake

```python
sample = "L12TO1"
def get_ADT_path(run):
    '''
    path getter
    '''
    return f"ADT_files/ADT_{run}.csv"

def feature_CLI(sample):
    '''
    adds the --feature ref [ref_path] option to the cellranger command line if Antibody Capture is part of the df for that sample
    '''
    
    if "Antibody Capture" in list(df.loc[sample, "library_type"]):
        run = df.loc[sample, "Run"][0]
        adt_file = get_ADT_path(run)
        return f" --feature-ref={adt_file}"
    return ""
```
